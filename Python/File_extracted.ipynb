```python
import os
import gzip
import shutil
```

```python
main_dir = "/Users/echandrasekara2/Documents/Kobza/Kobza Data"
output_dir = os.path.join(main_dir, "extracted_fastqs")
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
for root, _, files in os.walk(main_dir):
    for file in files:
        if file.endswith(".fastq.gz"):
            input_path = os.path.join(root, file)
            output_filename = file[:-3]  # remove '.gz'
            output_path = os.path.join(output_dir, output_filename)

            # Extract only if not already extracted
            if not os.path.exists(output_path):
                with gzip.open(input_path, 'rb') as f_in:
                    with open(output_path, 'wb') as f_out:
                        shutil.copyfileobj(f_in, f_out)
                print(f"Extracted: {output_filename}")
            else:
                print(f"Skipped (already exists): {output_filename}")
                
```
