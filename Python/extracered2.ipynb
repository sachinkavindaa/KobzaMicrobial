```python
import os
import shutil

# Define your main data directory
main_dir = "/Users/echandrasekara2/Documents/Kobza/Kobza Data Files"

# Define the destination folder for all fastq.gz files
output_dir = os.path.join(main_dir, "all_fastq_gz")
os.makedirs(output_dir, exist_ok=True)

# Recursively walk through the directory to find and move .fastq.gz files
for root, _, files in os.walk(main_dir):
    for file in files:
        if file.endswith(".fastq.gz"):
            source_path = os.path.join(root, file)
            dest_path = os.path.join(output_dir, file)

            # Move only if not already present
            if not os.path.exists(dest_path):
                shutil.move(source_path, dest_path)
                print(f"Moved: {file}")
            else:
                print(f"Skipped (already in folder): {file}")

print(f"\n✅ All .fastq.gz files moved to:\n{output_dir}")
```
